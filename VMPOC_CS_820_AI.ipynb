##Problem:
The Multi Objective Optimization suggests that, there should not be any compromize with any features of the system in order to achive the better performance or more feasible solution. Here we have taken a real life scenario where we have to place the Virtual Machines(VMs) to Cloud environment in an effecrive manner while satisfying following properties: 
### 1) Minimizing Power Consumption 
### 2) Minimizing Resource Wastage



```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import os
import datetime
from pandas import DataFrame
import collections 
import itertools 
import ast  
import math
import random
```

```
# Defining The Parameters
Server_quantity = input("Enter the No of Servers: ")
VM_quantity = input("Enter the No. of Virtual Machines: ")
```

```
Server = []
VMS =[]
i = 0
while(i < int(Server_quantity)):
  Server.append("S"+str(i))
  i+=1
print("Severs: ",Server)
# VMS
j = 0
while(j < int(VM_quantity)):
  VMS.append("VM"+str(j))
  j+=1
print("Virtual Machines: ",VMS)
```

```
# server thresold
Server_Processor_Thresold = 85
Server_Memory_Thresold = 90

# Virtual machine requirements
VM_Memory_Requirement= []
VM_CPU_Requirement = []
for v in VMS:
  VM_Memory_Requirement.append(random.randint(15,80))
  VM_CPU_Requirement.append(random.randint(5,80))

#Server Requirements
Server_Power_Consumption =[]
Server_power_consumption_busy = 215
Server_power_consumption_idle = 162
Server_Processor_utilization_list = []
Server_Processor_Thresold_list = []
Server_Memory_Thresold_list = []
Server_Memory_utlization_list = []
Server_Resource_Wastage = []
# Power Consumption Modelling
for i in Server:
  Server_processor_utlization = random.randint(0,100)
  Server_Memory_utlization = random.randint(0,100)
  if Server_processor_utlization !=0:
    Server_Power_Consumption.append(round((Server_power_consumption_busy-Server_power_consumption_idle)*Server_processor_utlization + Server_power_consumption_idle,2))
  else:
    Server_Power_Consumption.append(0)
  Server_Memory_utlization_list.append(Server_Memory_utlization)
  Server_Processor_utilization_list.append(Server_processor_utlization)
  Server_Memory_Thresold_list.append(Server_Memory_Thresold)
  Server_Processor_Thresold_list.append(Server_Processor_Thresold)

Server_normalized_processor_usage = [i/max(Server_Processor_utilization_list) for i in Server_Processor_utilization_list]
Server_normalized_memory_usage = [i/max(Server_Memory_utlization_list) for i in Server_Memory_utlization_list]

print(len(Server_Memory_utlization_list))
print(len(Server_normalized_memory_usage))
print(len(Server_normalized_processor_usage))
# Resource wastage modelling
for i in range(len(Server)):
  Server_Resource_Wastage.append(abs((Server_normalized_processor_usage[i]-Server_normalized_memory_usage[i])/(Server_Processor_utilization_list[i]-Server_Memory_utlization_list[i])))
```

```
vmpoc_df = pd.DataFrame(Server, columns =['Servers'])
vmpoc_df['VMs'] = VMS
vmpoc_df['VM Memory Req.'] = VM_Memory_Requirement
vmpoc_df['VM CPU Req.'] = VM_CPU_Requirement
vmpoc_df['Server CPU Thresold'] = Server_Processor_Thresold_list
vmpoc_df['Server Memory Thresold'] = Server_Memory_Thresold_list
vmpoc_df['Server Power Consumption'] = Server_Power_Consumption
vmpoc_df['Server Power Utilization'] = Server_Processor_utilization_list
vmpoc_df['Server Resource Wastage'] = Server_Resource_Wastage
vmpoc_df
```

```
no_of_ants = 1
Ants =[]
i = 0
while(i < no_of_ants):
  Ants.append("A"+str(i))
  i+=1
print("Ants: ",Ants)
Pareto_Optimal_Solution = []
pheromone_trails = [[pheromone_trail_initializer for i in range(len(Server))] for j in range(len(VMS))]
print(pheromone_trails)
```

```
VMS_temp = VMS
Server_temp = Server
assignments = {}
server_assign_status = {}

vm_memory_req_dict = {}
vm_memory_req_dict = {VMS[i]:VM_Memory_Requirement[i] for i in range(len(VMS))}

vm_cpu_req_dict = {}
vm_cpu_req_dict = {VMS[i]: VM_CPU_Requirement[i] for i in range(len(VMS))}

for ant in Ants:
  random.shuffle(Server_temp)
  random_server_init = Server_temp
  # print("hit here")
  while len(VMS_temp)>0:
    server_limit_flag = 0
    remain_vms = []
    server_remain_processor = Server_Processor_Thresold
    server_remain_memory = Server_Memory_Thresold
    for vm in VMS_temp:
      print("Chosen vm:",vm)
      print("Choosen Server: ",random_server_init[0])
      print("----")
      print("Server Memory:",server_remain_memory)
      print("Sever CPU:",server_remain_processor)
      print("VM Memory Req:",vm_memory_req_dict.get(vm))
      print("VM CPU Req:", vm_cpu_req_dict.get(vm))
      if (server_remain_memory >= vm_memory_req_dict.get(vm)) and (server_remain_processor >= vm_cpu_req_dict.get(vm)):
        server_remain_memory -=vm_memory_req_dict.get(vm)
        print("Remain Memory after assignment:",server_remain_memory)
        server_remain_processor -= vm_cpu_req_dict.get(vm)
        print("Remsin CPU after assignment:",server_remain_processor)
        assignments[random_server_init[0]] = vm
        VMS_temp.remove(vm)
        print("Remaining VMS:",VMS_temp)
      else:
        print("Adding to Remaining VM list")
        remain_vms.append(vm)
    print("All Remaining VMS: ",remain_vms)
  
```
